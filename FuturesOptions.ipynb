```python
import numpy as np
import matplotlib.pylab as plt
```

```python
import sop
```

```python
kl = np.linspace(50, 150)
sl = np.linspace(50, 150)
spot = 100
sigma = 0.25
T = 0.25
rf = 0.4
```

```python
c = [sop.BSM(si, 100, T, rf,sigma, 'C') for si in sl]
c_lr = [sop.LRBinomial(si, 100, T, rf, sigma, 20, 'C') for si in sl]
```

```python
plt.plot(sl, c)
plt.plot(sl, c_lr)
```

```python
plt.plot(sl, [np.log(np.abs(c_lr[i] - c[i])) for i in range(len(c))])
```

```python
p = [sop.BSM(si, 100, T, rf, sigma, 'P') for si in sl]
p_lr = [sop.LRBinomial(si, 100, T, rf, sigma, 20, 'P') for si in sl]
```

```python
plt.plot(sl, [np.log(np.abs(p_lr[i] - p[i])) for i in range(len(p))])
```

```python
plt.plot(sl, p)
plt.plot(sl, p_lr)
```
